```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 12})

from customRead import bigFile, smallFile, oneBodyFile
```

```python
filename = "../Output/EnergyPerAlpha/EnergyPerAlpha_Grid.txt"

alpha, E, E2, VAR, accepted, t1 = smallFile(filename)

def Eanalytic(alpha):
    return 1.5 * alpha + 3 / (8 * alpha)


plt.plot(alpha, E/10, label = "VMC")
plt.plot(alpha, Eanalytic(alpha), linestyle="--", label = "Analytic")
plt.xlabel(r"$\alpha$")
plt.ylabel(r"Energy per particle")
plt.title(r"Energy as a function of $\alpha$")
plt.legend()
plt.show()
```

```python
for i, lr in zip([0, 1, 2, 3], [0.001, 0.01, 0.1, 0.2]):
    filename = f"../Output/EnergyPerAlpha/EnergyPerLr{i}_GD.txt"
    
    alpha, E, E2, VAR, accepted, t1 = smallFile(filename)

    plt.plot(alpha, E/10)
    plt.title(f"Gradient descent with learning rate {lr}")
    plt.xlabel(r"$\alpha$")
    plt.ylabel(r"Energy per particle")
    plt.show()
```
