```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from customRead import bigFile, smallFile, oneBodyFile
```

```python
a_r_list = []
a_density_list = []

for a in ["0.000000", "0.004430", "0.044300"]:
    filename = f"../Output/Different_a_vals/a_loop_finaltry{a}_oB.txt"
    
    r, density = oneBodyFile(filename)

    a_r_list.append(r)
    a_density_list.append(density/np.sum(density))
    
plt.title(r"One Body Density $\rho(r)$ for Varying Correlation Strength")
plt.plot(a_r_list[0], a_density_list[0], marker = ".", color = "black", label = "a = 0.0")
plt.plot(a_r_list[1], a_density_list[1], 'bo', label = "a = 0.00443")
plt.plot(a_r_list[2], a_density_list[2],'ro', label = "a = 0.0443")

plt.xlabel(r"r"); plt.ylabel(r"$\rho(r)$")
plt.grid()
plt.legend()
plt.show()

plt.show()
```
