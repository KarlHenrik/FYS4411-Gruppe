## Testing scipy differential equation solver

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import complex_ode
```

```python
y0 = [2, 2]

t0 = 0
t1 = 10
dt = 0.001

def f(t, y):
    return [np.sin(t), -y[1]]
```

```python
n = int((t1 - t0) / dt) + 2

t = np.zeros(n)
t[0] = t0
y = np.zeros((n, 2), dtype=complex)
y[0] = y0
```

```python
r = complex_ode(f).set_integrator('vode')
r.set_initial_value(y0, t0)

i = 1
while r.successful() and r.t < t1:
    t[i] = r.t + dt
    y[i] = r.integrate(r.t + dt)
    i += 1
    
t = t[:i]
y = y[:i]
```

```python
plt.plot(t, y[:, 0].real)
plt.show()
```

```python
l = 6
#C = np.zeros((l, l), dtype="c16")
C = np.zeros((l, l))
for i in range(0, l, 2):
    vals = (np.random.rand(l // 2) - 0.5)*2
    up = np.zeros(l)
    down = np.zeros(l)
    for j in range(l // 2):
        up[2 * j] = vals[j]
        down[2 * j + 1] = vals[j]
    C[:, i] = up / np.linalg.norm(up)
    C[:, i + 1] = down / np.linalg.norm(down)
```

```python
def overlapd(C, D):
    
    l = C.shape[0]
    lap1 = 0
    lap2 = 0
    lap3 = 0
    lap4 = 0
    
    for a in range(l):
        lap1 += np.conj(C[a, 0]) * D[a, 0]
    for a in range(l):
        lap2 += np.conj(C[a, 1]) * D[a, 1]
    for a in range(l):
        lap3 += np.conj(C[a, 0]) * D[a, 1]
    for a in range(l):
        lap4 += np.conj(C[a, 1]) * D[a, 0]
    return np.abs(lap1 * lap2 - lap3 * lap4)**2
```
